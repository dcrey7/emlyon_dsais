Add a color scale representing the number of data we have for each point.

```python
scatter_df1=df[df['Season']=="Summer"].groupby("Sport").agg({"Height(cm)":"mean","Weight(kg)":"mean","Age":"count"}).reset_index()
scatter_df1.rename(columns={"Age":"Count"},inplace=True)
scatter_df1.dropna(inplace=True)
scatter_df1.reset_index(inplace=True)
scatter_df1
```

```python
x=scatter_df1['Height(cm)']
y=scatter_df1['Weight(kg)']
sportname=list(scatter_df1["Sport"])
count_list=list(scatter_df1['Count'])


plt.figure(figsize=(10,10))

plt.scatter(x, y, c=count_list, cmap="jet")
#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x,y, 1)

y_jittered=[i+random.uniform(-0.5,0.5) for i in y] 

#add linear regression line to scatterplot 
plt.plot(x, m*x+b)

# Add labels
for i in range(len(x)): 
    plt.annotate((sportname[i]), (x[i], y_jittered[i] ),fontsize=7)
```

__Question 3. (bonus)__
Choose 3 sports and represent the evolution of average weight over average height over the years.

Represent them first on 3 different graphs and then on the same one with different colors.

```python
athelitcs_df=df[df['Sport']=="Athletics"].groupby("Year")[["Height(cm)","Weight(kg)"]].mean().reset_index()
fencing_df=df[df['Sport']=="Fencing"].groupby("Year")[["Height(cm)","Weight(kg)"]].mean().reset_index()
shooting_df=df[df['Sport']=="Shooting"].groupby("Year")[["Height(cm)","Weight(kg)"]].mean().reset_index()
```

```python
athelitcs_df.head()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(10,2))
fig.tight_layout()
ax[0].plot(athelitcs_df['Year'],athelitcs_df['Height(cm)'],color="green",label="height")
ax[0].plot(athelitcs_df['Year'],athelitcs_df['Weight(kg)'],color="blue",label="weight")
ax[0].set_title("Athletics")
leg = ax[0].legend(loc="center right")

ax[1].plot(fencing_df['Year'],fencing_df['Height(cm)'],color="green",label="height")
ax[1].plot(fencing_df['Year'],fencing_df['Weight(kg)'],color="blue",label="weight")
ax[1].set_title("Fencing")
leg = ax[1].legend(loc="center right")


ax[2].plot(shooting_df['Year'],shooting_df['Height(cm)'],color="green",label="height")
ax[2].plot(shooting_df['Year'],shooting_df['Weight(kg)'],color="blue",label="weight")
ax[2].set_title("Shooting")
leg = ax[2].legend(loc="center right")

```
