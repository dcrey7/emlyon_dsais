## Part 2: Modifying the data

__Question 2.__
The column Games does not interest us since the information is already contained in the columns Year and Season: remove it. Make sure that your dataframe is updated.

```python
olympics = olympics.drop(columns='Games')

olympics
```

__Question 3.__
We want to add a column that directly tells us wether the athlete has won a medal or not. For this we will rename the Medal column 'Medal_type' and add a new column named 'Medal' with this information.  

```python
#First, rename the original Medal column. 
#Make sure that your dataframe is updated.

olympics = olympics.rename(columns={'Medal':'Medal_type'})
olympics
```

```python
#Then we will add the new Medal column.

#Find the different values that exist in the Medal_type column (tip: use the function .unique())
#Then write a function that returns true when the athlete won a medal (tip: use logic operators (Session 1).
print(olympics['Medal_type'].unique())

def has_medal(x):
    return (x == "Gold" or x == "Silver" or x=="Bronze")

#Create the new column thanks to the function.
#Don't forget to verify it worked correctly, by retrieving the new column again! 
#(tip: use the function .drop_duplicate())
olympics['Medal'] = olympics['Medal_type'].apply(has_medal)

olympics[['Medal_type','Medal']].drop_duplicates()
```

## Part 4: Plotting the data

__Question 1.__
- Take the result of the previous question and display a scatter plot representing the average weight as a function of the average height. (That means we statistically assume that height "causes" weight. When plotting, that means put height on the horizontal or X-axis and weight on the Y-axis). Don't take into account NaN values. (tip: use the dropna() method for this).

- Does theses values seem to be correlated? Is there some outlier sport? Don't forget to put a title and axes labels to your figure.

__(bonus)__ Print a regression line on top of your graph. Change its color so we can see it well. (tip: search numpy library to get a function for the regression).

__(bonus)__ Change the color and add the labels with the name of sport for the outliers points.

```python
import matplotlib.pyplot as plt

avg_height_weight = olympics[olympics["Season"]=="Summer"].groupby('Sport')[['Height (cm)', 'Weight (kg)']].mean().sort_values(by=['Height (cm)'])
avg_height_weight.dropna(inplace=True)

h = avg_height_weight['Height (cm)'].tolist()
w = avg_height_weight['Weight (kg)'].tolist()

fig, ax = plt.subplots(nrows=1, ncols=1)

ax.scatter(h,w)
ax.set_title("Average weight in function of averag height")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")

plt.show()
```
