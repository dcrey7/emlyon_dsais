### Exercise 1: read in a text file

Download from Brightspace the file StudentListfirstnames.txt. Put it in a folder you call 'data', next to where you keep your notebooks. Open with python and read the content into a list, each element of which should contain one student name. Remember to close the file after it is read. Print the list to make sure it has gone well.

```python
# Read the content into a list, each element of which should contain one student name.

my_file = open(file='../data/StudentlistFirstnames.txt', mode='r')
student_names = my_file.read().splitlines()
my_file.close()
print(student_names)
```

Two names are missing! Those of the teachers.
`teacher_names = ['Saeed', 'Vincent']`
In fact, there may be last-minute changes to the student list, so there may be more names to add.

Add the missing names to your file using the write functionality, and make your code robust:
if the names have already been entered into the file, they should not be added again when the same script is rerun. A good way may be to first add them to your list here in Python, then write.

```python
# Add the missing names to the list if they are not already there
teacher_names = ['Saeed', 'Vincent']

for teacher_name in teacher_names:
    if teacher_name not in student_names:
        student_names.append(teacher_name)

# Write the updated list to the file
my_file = open(file='../data/StudentlistFirstnames.txt', mode='w')
for student_name in student_names:
    my_file.write(student_name + '\n')
my_file.close()
```

We are going to make a function to calculate the numerological value of your name. (The expression number mentioned on wikihow, using all letters).

First think of the best Python object (iterable) to hold the mapping of the letters-to-values, and implement it. Tip: it may be best to map only capital letters in this object. You can always convert things later.

```python
letter_values = {
    'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9,
    'J': 1, 'K': 2, 'L': 3, 'M': 4, 'N': 5, 'O': 6, 'P': 7, 'Q': 8, 'R': 9,
    'S': 1, 'T': 2, 'U': 3, 'V': 4, 'W': 5, 'X': 6, 'Y': 7, 'Z': 8
}
```

(Please don't spend more than 3 minutes looking into the interpretation online. The next step actually interprets your number. In designing these materials, I aim to make them enjoyable, but I need your cooperation to keep them productive also.)

```python

def calculate_string_value(string, letter_values):
    # Convert the string to uppercase and calculate the total value
    return sum([letter_values.get(char.upper(), 0) for char in string])

def reduce_to_single_digit(n):
    while n >= 10 and n not in [11, 22, 33]:
        n = sum(int(digit) for digit in str(n))
    return n

def get_numerology(string, letter_values):
    return reduce_to_single_digit(calculate_string_value(string, letter_values))

# Example usage:
string = "Hello"
total_value = get_numerology(string, letter_values)
print(f"The total value of '{string}' is: {total_value}")
```

### Exercise 3b:
Below is the list with qualities per number from the wikihow article, as a dictionary.
Those who have 33, since you are not represented in the article you can remain mysterious.


```python
nums_quals = {
    1: ["Initiator of action", "pioneering", "leading", "independent", "attaining", "individualistic"],
    2: ["Cooperation", "adaptability", "consideration of others", "partnering", "mediating"],
    3: ["Expression", "verbalization", "socialization", "the arts", "the joy of living"],
    4: ["Values foundation", "order", "service", "struggle against limits", "steady growth"],
    5: ["Expansiveness", "visionary", "adventure", "the constructive use of freedom"],
    6: ["Responsibility", "protection", "nurturing", "community", "balance", "sympathy"],
    7: ["Analysis", "understanding", "knowledge", "awareness", "studious", "meditating"],
    8: ["Practical endeavors", "status-oriented", "power-seeking", "high-material goals"],
    9: ["Humanitarian", "giving nature", "selflessness", "obligations", "creative expression"],
    11: ["Higher spiritual plane", "intuitive", "illumination", "idealist", "a dreamer"],
    22: ["The Master Builder", "large endeavors", "powerful force", "leadership"],
    33: []
}
```

Now use the above dictionary in code to get your reading. Example expected output (replacing Vincent with whichever name was inputted of course):

`Dear Vincent, these are your adjectives:
Responsibility, protection, nurturing, community, balance, sympathy.`

If your number is 33, expected output is:

`Dear Vincent, you are truly a mystery.`

```python
def get_qualitative_reading(number):
    
    if number == 33:
        return "you are truly a mystery."
    else:
        qualities = nums_quals[number]
        return "these are your adjectives:\n" + ", ".join(qualities[:-1]) + ", " + qualities[-1] + "."

def get_my_reading(name, letter_values):
    number = get_numerology(name, letter_values)
    qualities = get_qualitative_reading(number)
    print(f"Dear {name}, {qualities}")

string = "Vincent"
get_my_reading(string, letter_values)
```

### Exercise 4d:

Copy your code from the last attempt, and ajust two things: first add a small amount of random jitter to the y-coordinates, using numpy's random.uniform() function, and set the low and high limits to -0.5, 0.5.

Now make a colors array by again using numpy's random.rand(). The first dimension should be as long as your number of points, the other should be 3: the RGB code (see today's Session).

Now when you call plt.scatter(), use your color list as the c argument.

Finally, we will use adjust_text to move the annotations. Try to understand all the input arguments.

```python
np.random.seed(42)

# Add some random noise to the y-coordinates
y_jittered = y + np.random.uniform(-0.5, 0.5, size=len(y))

# Create a scatterplot with random colors
plt.figure(figsize=(10, 6))
colors = np.random.rand(len(x), 3)  # Generate random RGB colors
plt.scatter(x, y_jittered, c=colors)

# Add labels
texts = []
for i, name in enumerate(student_names):
    text = plt.annotate(name, (x[i], y_jittered[i]), fontsize=9, ha='right')
    texts.append(text)

# Adjust the text to prevent overlapping
adjust_text(texts, x=x, y=y_jittered, arrowprops=dict(arrowstyle="->", color='k', lw=0.5))

plt.xlabel('Value')
plt.ylabel('Letter Count')
plt.title('Scatterplot of Value vs Letter Count')
plt.show()
```
