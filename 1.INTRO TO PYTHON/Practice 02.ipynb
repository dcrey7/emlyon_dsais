### Exercise 1: read in a text file

Download from Brightspace the file StudentListfirstnames.txt. Put it in a folder you call 'data', next to where you keep your notebooks. Open with python and read the content into a list, each element of which should contain one student name. Remember to close the file after it is read. Print the list to make sure it has gone well.

```python
# Read the content into a list, each element of which should contain one student name.
student_names=[]
with open(r"data/StudentListfirstnames.txt",'r') as myfile:
    lines=myfile.readlines()
    for names in lines:
        student_names.append(names.strip("\n"))
        
print(student_names)
```

Two names are missing! Those of the teachers.
`teacher_names = ['Saeed', 'Vincent']`
In fact, there may be last-minute changes to the student list, so there may be more names to add.

Add the missing names to your file using the write functionality, and make your code robust:
if the names have already been entered into the file, they should not be added again when the same script is rerun. A good way may be to first add them to your list here in Python, then write.

```python
# Add the missing names to the list if they are not already there
teacher_names = ['Saeed', 'Vincent']
for teacher_name in teacher_names:
    if teacher_name not in student_names:
        with open(r"data/StudentListfirstnames.txt",'a') as myfile:
            
            myfile.write(teacher_name+"\n"  )
            
        

# Write the updated list to the file
student_names=[]
with open(r"data/StudentListfirstnames.txt",'r') as myfile:
    lines=myfile.readlines()
    for names in lines:
        student_names.append(names.strip("\n"))
        
print(student_names)
```

```python
letter_values = {
    'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9,
    'J': 1, 'K': 2, 'L': 3, 'M': 4, 'N': 5, 'O': 6, 'P': 7, 'Q': 8, 'R': 9,
    'S': 1, 'T': 2, 'U': 3, 'V': 4, 'W': 5, 'X': 6, 'Y': 7, 'Z': 8
}
```

```python
#1

#2

#3
def get_numerology(string, letter_values):
    number_value=[]
    for i in string:
        for j in letter_values.keys():
            if i.upper()==j:
                number_value.append(letter_values[j])
    sum_number=sum(number_value)
    digit1=sum_number//10
    digit2=sum_number%10
    reduced_number=digit1+digit2
    return reduced_number
                
    
```

```python
# Example usage:
string = "Hello"
total_value = get_numerology(string, letter_values)
print(f"The total value of '{string}' is: {total_value}")
```

Now use the above dictionary in code to get your reading. Example expected output (replacing Vincent with whichever name was inputted of course):

`Dear Vincent, these are your adjectives:

If your number is 33, expected output is:

```python
def get_qualitative_reading(number):
    for j in nums_quals.keys():
        if number==j:
            quality=nums_quals[j]
        elif number==33:
            quality="mystery"
    return quality
            
    

def get_my_reading(name, letter_values):
    number = get_numerology(name, letter_values)
    quality= get_qualitative_reading(number)
    return quality

string = "Vincent"
quality=get_my_reading(string, letter_values)
quality
```
