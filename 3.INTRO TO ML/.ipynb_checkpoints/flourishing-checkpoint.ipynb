### Categorical

```python
def msds_summary_table_full(data):
    if isinstance(data,pd.Series):
        data=data.tolist()
    output={}
    for i in data:
        output[i]={'cpt':data.count(i),'freq':(data.count(i)/len(data))*100}
    return output


def msds_plot_summary_table_pareto(my_dict,m):
    # Extracting keys and values (frequencies)
    
    keys = [str(k) for k in my_dict.keys()]
    vals = [my_dict[k]['freq'] for k in my_dict]
    
    # Sorting the dictionary by values in descending order
    my_dict_sorted = dict(zip(keys, vals))
    my_dict_sorted = {k: v for k, v in sorted(my_dict_sorted.items(), key=lambda item: item[1], reverse=True)}
    
    # Preparing cumulative sum values for the Pareto line
    cum_sum = 0
    cum_vals = []
    keys_sorted = [str(k) for k in my_dict_sorted.keys()]
    vals_sorted = [v for v in my_dict_sorted.values()]
    
    for i in vals_sorted:
        cum_sum += i
        cum_vals.append(cum_sum)
    
    # Normalize cumulative values to match the y-axis scale
    cum_vals_percent = [val / cum_sum * 100 for val in cum_vals]
    
    # Plotting the bar plot and the Pareto line
    fig, ax1 = plt.subplots(figsize=(8, 4))
    
    # Bar plot
    ax1.bar(range(len(keys_sorted)), vals_sorted, color='blue', alpha=0.7)
    ax1.set_xlabel("Keys")
    ax1.set_ylabel("Frequencies", color='blue')
    ax1.set_xticks(range(len(keys_sorted)))
    ax1.set_xticklabels(keys_sorted, rotation=45)
    
    # Pareto line plot
    ax2 = ax1.twinx()  # Create a second y-axis for cumulative percentage
    ax2.plot(range(len(keys_sorted)), cum_vals, '-o', color='red')
    ax2.set_ylabel("Cumulative frequency", color='red')
    
    plt.title(f"Pareto Chart for {m}")
    plt.tight_layout()
    plt.show()

    
```

```python
for i in categorical_cols:
    
    summary=msds_summary_table_full(df[i])
    msds_plot_summary_table_pareto(summary,i)
    
```

### Continous

```python
def msds_cum_histogram(data, N=5):
    data_min = min(data)
    data_max = max(data)
    rg = data_max - data_min
    bin_indices = [
        int(np.floor(N * (x - data_min) / rg))
        for x in data
        if x is not None and not np.isnan(x)
    ]
    n = len(bin_indices)
    bins_width = rg / N
    bins_lower = [data_min + bins_width * (i - 1) for i in range(1, N + 1)]
    bins_mid = [data_min + bins_width * (i - 0.5) for i in range(1, N + 1)]
    bins_upper = [data_min + bins_width * (i) for i in range(1, N + 1)]
    bins_data = np.zeros(N).astype("int32")
    for i in bin_indices:
        bins_data[min(i, N - 1)] += 1
    ds = []
    cum_data = np.cumsum(bins_data)
    for i in range(N):
        ds.append(
            {
                "lower": bins_lower[i],
                "upper": bins_upper[i],
                "mid": bins_mid[i],
                "count": bins_data[i],
                "freq": bins_data[i] / n * 100,
                "cum_freq": cum_data[i] / n * 100,
            }
        )
    return ds




def msds_plot_histogram_pareto(hist_data,m, title=None):
    x = ["{:.0f}".format(a["mid"]) for a in hist_data]
    y = [x["count"] for x in hist_data]
    cum_y = [x["cum_freq"] for x in hist_data]
    yLabel = "Count"

    plt.figure()
    sb.barplot(x=x, y=y, color=sb.color_palette()[0]).set(
        ylabel=yLabel,
        xlabel=title,
    )
    if len(hist_data) > 10:
        plt.xticks(rotation=45)
    ax2 = plt.twinx()
    sb.lineplot(x=x, y=cum_y, marker="o", color=sb.color_palette()[1]).set(
        ylabel="Cummulative freq (percent)"
    )
    plt.title(f"Pareto histogram Chart for {m}")
```

```python
for i in numerical_cols:
    hist_data=msds_cum_histogram(df[i])
    msds_plot_histogram_pareto(hist_data, i)
```
