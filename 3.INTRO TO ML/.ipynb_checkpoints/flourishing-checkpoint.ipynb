```python
def msds_summary_table_full(data):
    if isinstance(data, pd.Series):
        data = data.tolist()
    output = {}
    for i in data:
        output[i] = {'cpt': data.count(i), 'freq': (data.count(i) / len(data)) * 100}
    return output

def msds_plot_summary_table(my_dict, m):
    # Convert dictionary to DataFrame
    df = pd.DataFrame([(k, v['freq']) for k, v in my_dict.items()], columns=['Category', 'Frequency'])
    df = df.sort_values('Frequency', ascending=False)

    # Set up the matplotlib figure
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

    # Generate a color palette
    colors = plt.cm.Set3(np.linspace(0, 1, len(df)))

    # bar plot
    sns.barplot(x='Category', y='Frequency', data=df, ax=ax1, palette=colors)
    ax1.set_xlabel(m)
    ax1.set_ylabel("Frequencies")
    ax1.set_title(f"Pareto Chart for {m}")
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

    # Pie Chart
    ax2.pie(df['Frequency'], labels=df['Category'], autopct='%1.1f%%', startangle=90, colors=colors)
    ax2.set_title(f"Pie Chart for {m}")

    # Adjust layout and display plot
    plt.tight_layout()
    plt.show()
```

```python
for i in categorical_cols:
    
    summary=msds_summary_table_full(df[i])
    msds_plot_summary_table(summary,i)
    
```

### Continous

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def msds_cum_histogram(data, N=5):
    data_min = min(data)
    data_max = max(data)
    rg = data_max - data_min
    bin_indices = [
        int(np.floor(N * (x - data_min) / rg))
        for x in data
        if x is not None and not np.isnan(x)
    ]
    n = len(bin_indices)
    bins_width = rg / N
    bins_lower = [data_min + bins_width * (i - 1) for i in range(1, N + 1)]
    bins_mid = [data_min + bins_width * (i - 0.5) for i in range(1, N + 1)]
    bins_upper = [data_min + bins_width * (i) for i in range(1, N + 1)]
    bins_data = np.zeros(N).astype("int32")
    for i in bin_indices:
        bins_data[min(i, N - 1)] += 1
    ds = []
    cum_data = np.cumsum(bins_data)
    for i in range(N):
        ds.append(
            {
                "lower": bins_lower[i],
                "upper": bins_upper[i],
                "mid": bins_mid[i],
                "count": bins_data[i],
                "freq": bins_data[i] / n * 100,
                "cum_freq": cum_data[i] / n * 100,
            }
        )
    return ds

def msds_plot_histogram_pareto(hist_data, m, title=None):
    # Sort hist_data by descending frequency (to get Pareto effect)
    # hist_data = sorted(hist_data, key=lambda x: x['count'], reverse=True)
    
    x = ["{:.0f}".format(a["mid"]) for a in hist_data]
    y = [x["count"] for x in hist_data]
    # cum_y = np.cumsum([x["freq"] for x in hist_data])  # Cumulative frequency should always increase

    sns.set_theme(rc={'figure.figsize':(16,6)})
    fig, axs = plt.subplots(ncols=2)

    # Plot the histogram with the Pareto line
    sns.barplot(x=x, y=y, ax=axs[0])
    if len(hist_data) > 10:
        axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=45)
    
    # Plot the cumulative line on the same axis
    # sns.lineplot(x=x, y=cum_y, ax=axs[0], marker="o", color=sns.color_palette()[1])
    axs[0].set_ylabel("Cummulative freq (percent)")
    axs[0].set_title(f"Pareto Chart for {m}")

    # Plot the boxplot
    axs[1].set_title(f"Box plot for {m}")
    sns.boxplot(df[m], ax=axs[1])

    # Show the plot
    plt.show()
```

```python
for i in numerical_cols:
    hist_data=msds_cum_histogram(df[i])
    msds_plot_histogram_pareto(hist_data, i)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

siginicant_list = []
non_significant_list = []
def bivariate_Categorical(x,y,a,b):
    contingency_table = pd.crosstab(df[x], df[y])
    print("Contingency Table:")
    print(contingency_table)
    

    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    print(f"\nChi-square statistic: {chi2:.4f}")
    print(f"p-value: {p_value:.4f}")
    print(f"Degrees of freedom: {dof}")
    
   
    alpha = 0.05  # Common significance level
    
    if p_value <= alpha:
        print(f"\nThe p-value ({p_value:.4f}) is less than or equal to the significance level ({alpha}).")
        print("We reject the null hypothesis.")
        print(f"There is a significant relationship between {x} and {y}.")
        a.append((contingency_table,x,y,p_value))
    else:
        print(f"\nThe p-value ({p_value:.4f}) is greater than the significance level ({alpha}).")
        print("We fail to reject the null hypothesis.")
        print(f"There is not enough evidence to conclude a significant relationship between {x} and {y}.")
        b.append((contingency_table,x,y,p_value))
    
    # Optional: Calculate and display Cramer's V for effect size
    def cramers_v(contingency_table):
        chi2 = chi2_contingency(contingency_table)[0]
        n = contingency_table.sum().sum()
        min_dim = min(contingency_table.shape) - 1
        return np.sqrt(chi2 / (n * min_dim))
    
    cramer_v = cramers_v(contingency_table)
    print(f"\nCramer's V: {cramer_v:.4f}")

```

```python


for i in range(len(categorical_cols)):
    for j in range(i + 1, len(categorical_cols)):  # Corrected the range
        bivariate_Categorical(categorical_cols[i], categorical_cols[j], siginicant_list, non_significant_list)
```

```python
for i, value in enumerate(siginicant_list):
    dfs = siginicant_list[i][0].reset_index()
    dfs1 = pd.melt(dfs, id_vars=value[i])
    
    # Print column names
    print("Columns in dfs1:", dfs1.columns)
    
    # Create a figure with two subplots side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    
    # Grouped bar chart
    sns.barplot(x=dfs1.columns[0], y='value', hue=dfs1.columns[1], data=dfs1, ax=ax1)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
    ax1.set_title('Grouped Bar Chart')
    
    # Stacked bar chart
    # Use the actual column names from dfs1
    pivot_data = dfs1.pivot(index=dfs1.columns[0], columns=dfs1.columns[1], values='value')
    pivot_data.plot(kind='bar', stacked=True, ax=ax2)
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
    ax2.set_title('Stacked Bar Chart')
    
    # Adjust layout and show plot
    plt.tight_layout()
    plt.show()
```
