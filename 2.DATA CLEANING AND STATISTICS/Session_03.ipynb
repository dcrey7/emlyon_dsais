```python
def msds_contingency_table(data1, data2, with_totals=True):
    data1 = list(data1)
    # because in salary test we drop nans and this cause index problems so to avoid index problems
    # we should transform the data fram into a list or we can elimintate the NaN at index 172 and
    # NaN at index 260 by df.reset_index(drop = True)
    data2 = list(data2)
    cols = sorted(list(set(data2)))
    cols.append("Total")
    rows = sorted(list(set(data1)))
    rows.append("Total")
    contingency = [[0 for i in cols] for j in rows]
    for i in range(len(data1)):
        x = rows.index(data1[i])  # find the row index of data1[i]
        y = cols.index(data2[i])
        contingency[x][y] += 1
        contingency[len(rows) - 1][y] += 1
        contingency[x][len(cols) - 1] += 1
        contingency[len(rows) - 1][len(cols) - 1] += 1
    df = pd.DataFrame(contingency, index=rows, columns=cols)
    return df


def test_msds_contingency_table():
    test_data1 = ["A", "A", "B", "C", "C", "A", "A"]
    test_data2 = ["Z", "Z", "X", "X", "Z", "X", "Z"]
    cont_table = msds_contingency_table(test_data1, test_data2)
    print(cont_table)
    assert cont_table.shape == (4, 3)
    assert cont_table.Z["A"] == 3
    # Compare with crosstab function
    print("CROSSTAB RESULTS")
    print(
        pd.crosstab(
            test_data1, test_data2, margins=True, colnames=["data1"], rownames=["data2"]
        )
    )

    print("SALARY TEST")
    salary = pd.read_csv("data/salary.csv")
    salary = salary.dropna()
    gender_ed = msds_contingency_table(salary.Gender, salary["Education Level"])
    print(gender_ed)
    print("CROSSTAB RESULTS")
    print(pd.crosstab(salary.Gender, salary["Education Level"], margins=True))


test_msds_contingency_table()
```

```python
def msds_contingency_table_overall(data1, data2, with_totals=True):
        df=msds_contingency_table(data1, data2, with_totals=True)
        df=df/len(data1)*100
        return df


def test_msds_contingency_table_overall():
    test_data1 = ["A", "A", "B", "C", "C", "A", "A"]
    test_data2 = ["Z", "Z", "X", "X", "Z", "X", "Z"]
    cont_table = msds_contingency_table_overall(test_data1, test_data2)
    print(cont_table)
    assert cont_table.shape == (4, 3)
    assert abs(cont_table.Total["Total"] - 100) < 1e-5
    print("CROSSTAB RESULTS")
    print(
        pd.crosstab(
            test_data1,
            test_data2,
            margins=True,
            normalize=True,
            colnames=["data1"],
            rownames=["data2"],
        )
        * 100
    )

    print("SALARY TEST")
    salary = pd.read_csv("data/salary.csv")
    salary = salary.dropna()
    gender_ed = msds_contingency_table_overall(salary.Gender, salary["Education Level"])
    print(gender_ed)
    print("CROSSTAB RESULTS")
    print(
        pd.crosstab(
            salary.Gender, salary["Education Level"], normalize=True, margins=True
        )
        * 100
    )


test_msds_contingency_table_overall()
```

### [easy] Covariance 

Write a function that computes the covariance of two quantitative data sets (of same size!).

The formula is as follows:

$\Large \text{cov}(X,Y) = \frac{\sum_{i=1}^{N}(X_i - \bar{X})(Y_i - \bar{Y})}{n-1}$

with:
 - $\bar{X}$ is the sample mean of X
 - $\bar{Y}$ is the sample mean of Y

Apply your function to the `salary.csv` dataset to see the correlation between age and experience.$
Validate your implementation by comparing your results with results from `numpy` package.

```python
def msds_covariance(data1, data2):
    sum_co=0
    x_mean=(sum(data1)/len(data1))
    y_mean=(sum(data2)/len(data2))
    for i in range(len(data1)):

        
        sum_co+=(data1[i]-x_mean)*(data2[i]-y_mean)
    return sum_co/(len(data1)-1)


def test_msds_covariance():
    cov = msds_covariance([1, 2, 3, 4], [1, 2, 3, 4])
    salary_data = pd.read_csv("data/salary.csv")
    salary_data = salary_data.dropna().reset_index(drop=True)  # Reset index after dropping NA values
    assert abs(np.cov([[1, 2, 3, 4], [1, 2, 3, 4]])[0][1] - cov) < 1e-5
    cov = msds_covariance(salary_data.Age, salary_data.Salary)
    assert abs(np.cov([salary_data.Age, salary_data.Salary])[0][1] - cov) < 1e-5


test_msds_covariance()
```

### [easy] Correlation coeficient 

Write a function that computes the correlation coeficient of two quantitative data sets (of same size!).

The formula is 

$\Large r = \frac{\text{cov}(X,Y)}{S_X S_Y}$

with:
 - $S_{X}$ is the sample variance of X
 - $S_{Y}$ is the sample variance of Y

Apply your function to the `salary.csv` dataset to see the correlation between age and experience.

Validate your implementation by comparing your results with results from `statistics` or `statmodels` packages.

```python
import numpy as np
import pandas as pd

def msds_sum(data):
    """
    Computes the sum of every element of the iterable
    """
    if len(data) == 0:
        return None
    return sum(data)

def msds_mean(data):
    """
    Computes the average of every element of the iterable
    """
    if len(data) == 0:
        return None
    return msds_sum(data) / len(data)

def msds_sample_variance(data):
    if len(data) < 2:
        return None
    mean = msds_mean(data)
    return sum((x - mean) ** 2 for x in data) / (len(data) - 1)

def msds_covariance(data1, data2):
    if len(data1) != len(data2) or len(data1) < 2:
        return None
    mean1, mean2 = msds_mean(data1), msds_mean(data2)
    return sum((x - mean1) * (y - mean2) for x, y in zip(data1, data2)) / (len(data1) - 1)

def msds_corr(data1, data2):
    if len(data1) != len(data2) or len(data1) < 2:
        return None
    cov = msds_covariance(data1, data2)
    var1 = msds_sample_variance(data1)
    var2 = msds_sample_variance(data2)
    if var1 is None or var2 is None or var1 == 0 or var2 == 0:
        return None
    return cov / (np.sqrt(var1) * np.sqrt(var2))

def test_msds_corr():
    # Test with simple data
    data1 = [1, 2, 3, 4]
    data2 = [1, 2, 3, 4]
    corr = msds_corr(data1, data2)
    np_corr = np.corrcoef(data1, data2)[0, 1]
    assert abs(np_corr - corr) < 1e-5, f"Expected {np_corr}, but got {corr}"

    # Test with salary data
    salary_data = pd.read_csv("data/salary.csv")
    salary_data = salary_data.dropna()
    age_list = salary_data.Age.tolist()
    salary_list = salary_data.Salary.tolist()
    corr = msds_corr(age_list, salary_list)
    np_corr = np.corrcoef(age_list, salary_list)[0, 1]
    assert abs(np_corr - corr) < 1e-5, f"Expected {np_corr}, but got {corr}"

    

test_msds_corr()
```
